```python
import torch
from prob_mbrl import envs, utils, models
```

```python
exp = utils.ExperienceDataset()
exp.load("/home/juancamilog/.prob_mbrl/results_DoubleCartpole_20181123112356.149824.pth.tar")

# initialize env
env = envs.DoubleCartpole()
env.dt = env.model.dt
D = env.observation_space.shape[0]
U = env.action_space.shape[0]
maxU = env.action_space.high

# initalize policy
pol_shape = [50]*2
pol_model = models.mlp(
        D, U, pol_shape,
        dropout_layers=[models.modules.BDropout(0.1)
                        for i in range(len(pol_shape))],
        nonlin=torch.nn.ReLU,
        weights_initializer=torch.nn.init.xavier_normal_,
        biases_initializer=None,
        output_nonlin=torch.nn.Tanh)

pol = models.Policy(pol_model, maxU).float()
```

```python
values = []
for it, pol_params in enumerate(exp.policy_parameters):
    if len(pol_params) == 0:
        continue
    print it
    for p, p_data in zip(pol.parameters(), pol_params):
        p.data  = p_data
    cb = lambda *args, **kwargs: env.render()
    values_it = []
    for i in range(10):
        ret = utils.apply_controller(env, pol, 60, callback=cb, realtime=False)
        values_it.append(sum(ret[2]))
        print values_it[-1]
    values.append(values_it)
values = torch.tensor(values)
```
